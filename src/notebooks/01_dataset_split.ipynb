```python
import pandas as pd
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
```

```python
RANDOM_STATE = 42
```

```python
df = pd.read_csv('Womens Clothing E-Commerce Reviews.csv')
df.head()
```

```python
(df.index == df['Unnamed: 0']).all()
```

ðŸ’¡ "Unnamed: 0" is actually review id

```python
df.rename(columns = {'Unnamed: 0': 'Review ID'}, inplace = True)
df.head()
```

```python
train_df, test_df = train_test_split(df, test_size=0.2, stratify=df['Rating'], random_state=RANDOM_STATE)
test_df, validation_df = train_test_split(test_df, test_size=0.5, stratify=test_df['Rating'], random_state=RANDOM_STATE)
```

```python
print(f'Train: {len(train_df)}, Validation: {len(validation_df)}, Test: {len(test_df)}, Sum: {len(train_df) + len(validation_df) + len(test_df)} == {len(df)}')
```

```python
plt.hist(train_df['Rating'])
plt.show()
```

```python
plt.hist(validation_df['Rating'])
plt.show()
```

```python
plt.hist(test_df['Rating'])
plt.show()
```

âš¡ Distributions between train, validation and test set are the same

```python
train_df.to_csv('train.csv', index=False)
validation_df.to_csv('validation.csv', index=False)
test_df.to_csv('test.csv', index=False)
```
